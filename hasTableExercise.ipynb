```python
temp_nyc={}

with open ("nyc_weather.csv") as f:
    next(f)  #skip header row
    for line in f:
        tokens=line.strip().split(",")
        day=tokens[0]
        temp=int(tokens[1])
        temp_nyc[day]=temp
```

```python
class HashTable:
    def __init__(self):
        self.MAX=10
        self.arr=[[] for i in range(self.MAX)]
    def get_hash(self,key):
        hash=0
        for i in key:
            hash+=ord(i)
        return hash%10
    def __setitem__(self,key,value):
        h=self.get_hash(key)
        found=False
        for idx,element in enumerate (self.arr[h]):
            if len(element)==2 and element[0]==key:
                self.arr[h][idx]=(key,value)
                found=True
                break
        if not found:
            self.arr[h].append((key,value))
    def __getitem__(self,key):
        h=self.get_hash(key)
        for el in self.arr[h]:
            if el[0]==key:
                return el[1]
            
        return self.arr[h]
    def __delitem__(self,key):
        h=self.get_hash(key)
        for idx,el in enumerate(self.arr[h]):
            if el[0]==key:
                del self.arr[h][idx]
                break
```
