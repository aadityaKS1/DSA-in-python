```python
import string
word_count={}
with open("poem.txt","r") as f:
    for line in f:
        tokens=line.split(" ")
        for token in tokens:
            token=token.strip(string.punctuation).lower()
            if token in word_count:
                word_count[token]+=1
            else:
                word_count[token]=1
                
```

```python
print(word_count)
```
