```python
# Hash Table using chaining to avoid collisions
class HashTable:
    def __init__(self):
        # Maximum size of hash table
        self.MAX = 10
        
        # Create an array of empty lists (buckets)
        # Each bucket will store (key, value) pairs
        self.arr = [[] for _ in range(self.MAX)]

    def get_hash(self, key):
        """
        Generate hash index by summing ASCII values
        of characters in the key and taking modulo.
        """
        hash_value = 0
        for ch in key:
            hash_value += ord(ch)
        return hash_value % self.MAX

    def __setitem__(self, key, value):
        """
        Insert or update a key-value pair.
        Uses chaining to handle collisions.
        """
        h = self.get_hash(key)
        found = False

        # Search bucket to check if key already exists
        for i, element in enumerate(self.arr[h]):
            if len(element) == 2 and element[0] == key:
                # Key found → update value
                self.arr[h][i] = (key, value)
                found = True
                break

        # Key not found → append new pair
        if not found:
            self.arr[h].append((key, value))

    def __getitem__(self, key):
        """
        Retrieve value for a given key.
        Returns None if key is not present.
        """
        h = self.get_hash(key)

        # Search in the bucket
        for element in self.arr[h]:
            if element[0] == key:
                return element[1]

    def __delitem__(self, key):
        """
        Delete key-value pair from the hash table.
        """
        h = self.get_hash(key)

        # Find the key and remove it
        for index, kv in enumerate(self.arr[h]):
            if kv[0] == key:
                del self.arr[h][index]
                break
```
